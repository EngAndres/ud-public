# EDA with ydata-profiling — `weather_problems.json`

**ydata-profiling** (formerly pandas-profiling) generates a full interactive HTML report from a DataFrame in a single call.

| Feature | What it shows |
|---|---|
| Overview | Dataset shape, warnings, duplicates |
| Variables | Distribution, missing %, distinct count, type |
| Interactions | Scatter matrix for numeric pairs |
| Correlations | Pearson, Spearman, Cramér's V |
| Missing values | Heatmap + bar chart of nulls |
| Samples | Head / tail rows |

> Dataset: Australian daily weather observations — 372 records with **intentional data quality issues** (missing values, duplicates, outliers).

## 00 · Install dependencies

```python
import importlib, subprocess, sys

if importlib.util.find_spec("ydata_profiling") is None:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "ydata-profiling"])
    print("✔ ydata-profiling installed")
else:
    import ydata_profiling
    print(f"✔ ydata-profiling already available (v{ydata_profiling.__version__})")
```

## 01 · Imports

```python
import pandas as pd
from pathlib import Path
from ydata_profiling import ProfileReport
import warnings

warnings.filterwarnings("ignore")
print("Imports OK")
```

## 02 · Load Data

```python
DATA_PATH = Path("weather_problems.json")

df = pd.read_json(DATA_PATH)
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
df.head(3)
```

```python
# Quick summary before profiling
print("=== dtypes ===")
print(df.dtypes)
print()
print("=== Missing values ===")
missing = df.isnull().sum()
print(missing[missing > 0].sort_values(ascending=False))
```

## 03 · Generate Profile Report

- `explorative=True` enables additional correlations and interactions
- `minimal=False` (default) produces the full report — set to `True` for large datasets (> 100k rows)

```python
profile = ProfileReport(
    df,
    title="Weather Dataset — ydata-profiling EDA",
    explorative=True,
    correlations={
        "pearson":  {"calculate": True},
        "spearman": {"calculate": True},
        "kendall":  {"calculate": False},
        "phi_k":    {"calculate": False},
        "cramers":  {"calculate": True},
    },
    missing_diagrams={
        "bar": True,
        "matrix": True,
        "heatmap": True,
    },
)
print("Profile generated.")
```

## 04 · Display Report (inline)

```python
profile.to_notebook_iframe()
```

## 05 · Save HTML Report

```python
out_path = Path("weather_profile_report.html")
profile.to_file(out_path)
print(f"Report saved → {out_path.resolve()}")
```
