```python
import pandas as pd

dataset_df = pd.read_csv('./StudentsInfo.csv')
dataset_df
```

## Clean Data



```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

columns = ['Company', 'Position']
features = dataset_df[columns]
target = dataset_df['Salary']

onehotencoder = OneHotEncoder(sparse_output=False)
feature_onehot = onehotencoder.fit_transform(features['Position'])
onehot_names = onehotencoder.get_feature_names_out(['Position'])
encoded_df = pd.DataFrame(feature_onehot, columns=onehot_names)
features = pd.concat([encoded_df, features['Company']], axis=1)

X_train, X_test, y_train, y_test = train_test_split(features, 
                                                    target, 
                                                    test_size=0.2, 
                                                    random_state=42,
                                                    stratify=target)
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error

xgboost_regressor = XGBRegressor(n_estimators=100, learning_rate=0.05)

#train
model = xgboost_regressor.fit(X_train, y_train)

#test
y_test_pred = model.predict(X_test)
error = mean_squared_error(y_test, y_test_pred)
```

```python
import pickle

with open("model.pkl", 'w') as f:
    pickle.dump(model, f)
```
