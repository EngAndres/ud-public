# Example 1: Euclidean GCD

This is an example of the __Euclides Algorithm__ to get the _greater common divisor_.

```python
def euclidean_gcd(a: int, b: int) -> int:
    """
    This function gets the GCD based on the euclidean algotirhm for this purpose.

    Args:
        a(int): first number
        b(int): second number

    Returns:
        An integer number to define the GCD for the provided numbers.
    """
    while b != 0:
        r = a % b
        a = b
        b = r
    return a

result = euclidean_gcd(9494949, 2932929)
print(result)
```

```python
def euclidean_gcd_recursive(a: int, b: int) -> int:
    """
    This function gets the GCD based on the euclidean algotirhm for this purpose.

    Args:
        a(int): first number
        b(int): second number

    Returns:
        An integer number to define the GCD for the provided numbers.
    """
    if b == 0:
        return a
    else:
        return euclidean_gcd_recursive(b, a % b)
    
result = euclidean_gcd_recursive(9494949, 2932929)
print(result)
```

# Example 2: Babylonean Method for Square Root

```python
def square_root(s: int) -> float:
    """
    This function calculates the square root of an integer number 
    based on Babylonean method.

    Args:
        s(int): the number to calculate the square root of

    Returns:
        A decimal number with the result of the square root approximation.
    """
    x = s /2
    DIFF = 0.001

    while True:
        x_next = (1 / 2) * (x + (s / x))
        diff_x = abs(x_next - x)
        x = x_next
        if diff_x <= DIFF:
            break

    return x

print( square_root(150) )
```

# Exercise 3: Find Maximum in an Array

```python
def find_maximum(array: list) -> int:
    """
    This function calculates the maxximum number in an integer array.
    If array is empty, an error message is returned.

    Args:
        array(list): the list of integers to find the maximum number.

    Returns:
        An integer number with the maximum value in the array.
    """
    if len(array) == 0:
        return "Cannot process and empty array"
    
    max = array[0]
    for i in range(1, len(array)):
        if array[i] > max:
            max = array[i]

    return max

print(find_maximum([11, 4, 6, 7, 2, 1, 9, 2, 9, 13]))
print(find_maximum([]))
```

# #4 Factorial

```python
def factorial(n):
    if n < 0:
        return "Error."
    result = 1
    for i in range(1, n+1):
        result = result * i
    return result

print(factorial(100))
```

```python
def factorial_recursive(n):
    if n == 0:
        return 1
    else:
        return n * factorial_recursive(n - 1)

print(factorial_recursive(100))
```

## Greedy: Coin Change

```python
def get_coin_change(n: int, denominations: list):
    denominations = sorted(denominations, reverse=True)
    result = []
    for d in denominations:
        if n >= d:
            coins = n // d
            result.append( (d, coins) )
            n -= (coins * d)
            if n == 0:
                break
    return result

denominations = [50, 200, 100, 1000, 500]

print( get_coin_change(40, denominations) )
```

## Queens

```python
data = input()

while data != "0 0 0 0": 
    x1, y1, x2, y2 = map(int, data.split(' '))
    result = None
    if x1 == x2 and y1 == y2:
        result = 0
    elif (x1 == x2) or (y1 == y2) or (abs(x1 - x2) == abs(y1 - y2)):
        result = 1
    else:
        result = 2
    print(result)
    data = input()
```

## Optical Reader

```python
data = int(input())
grade = ['A', 'B', 'C', 'D', 'E']

while data != 0:
    for i in range(data):
        answers = list(map(int, input().split(' ')))
        found = False
        winner = "*"    
        for j, answer in enumerate(answers):
            if answer <= 127:
                if not found:
                    winner = grade[j]
                    found = True
                else:
                    winner = "*"
        print(winner)

    data = int(input())
```
