# Example 1: Euclidean GCD

This is an example of the __Euclides Algorithm__ to get the _greater common divisor_.

```python
def euclidean_gcd(a: int, b: int) -> int:
    """
    This function gets the GCD based on the euclidean algotirhm for this purpose.

    Args:
        a(int): first number
        b(int): second number

    Returns:
        An integer number to define the GCD for the provided numbers.
    """
    while b != 0:
        r = a % b
        a = b
        b = r
    return a

result = euclidean_gcd(9494949, 2932929)
print(result)
```

```python
def euclidean_gcd_recursive(a: int, b: int) -> int:
    """
    This function gets the GCD based on the euclidean algotirhm for this purpose.

    Args:
        a(int): first number
        b(int): second number

    Returns:
        An integer number to define the GCD for the provided numbers.
    """
    if b == 0:
        return a
    else:
        return euclidean_gcd_recursive(b, a % b)
    
result = euclidean_gcd_recursive(9494949, 2932929)
print(result)
```

# Example 2: Babylonean Method for Square Root

```python
def square_root(s: int) -> float:
    """
    This function calculates the square root of an integer number 
    based on Babylonean method.

    Args:
        s(int): the number to calculate the square root of

    Returns:
        A decimal number with the result of the square root approximation.
    """
    x = s /2
    DIFF = 0.001

    while True:
        x_next = (1 / 2) * (x + (s / x))
        diff_x = abs(x_next - x)
        x = x_next
        if diff_x <= DIFF:
            break

    return x

print( square_root(150) )
```

# Exercise 3: Find Maximum in an Array

```python
def find_maximum(array: list) -> int:
    """
    This function calculates the maxximum number in an integer array.
    If array is empty, an error message is returned.

    Args:
        array(list): the list of integers to find the maximum number.

    Returns:
        An integer number with the maximum value in the array.
    """
    if len(array) == 0:
        return "Cannot process and empty array"
    
    max = array[0]
    for i in range(1, len(array)):
        if array[i] > max:
            max = array[i]

    return max

print(find_maximum([11, 4, 6, 7, 2, 1, 9, 2, 9, 13]))
print(find_maximum([]))
```

# #4 Factorial

```python
def factorial(n):
    if n < 0:
        return "Error."
    result = 1
    for i in range(1, n+1):
        result = result * i
    return result

print(factorial(100))
```

```python
def factorial_recursive(n):
    if n == 0:
        return 1
    else:
        return n * factorial_recursive(n - 1)

print(factorial_recursive(100))
```

## Greedy: Coin Change

```python
def get_coin_change(n: int, denominations: list):
    denominations = sorted(denominations, reverse=True)
    result = []
    for d in denominations:
        if n >= d:
            coins = n // d
            result.append( (d, coins) )
            n -= (coins * d)
            if n == 0:
                break
    return result

denominations = [50, 200, 100, 1000, 500]

print( get_coin_change(40, denominations) )
```

## N-Queens Problem with BackTracking

```python
def is_safe(board: list, n: int, row_q: int, col_q: int) -> bool:
    """
    This function evaluates if a new option is a valid
    partial solution, it means, a new queen is not 
    attacking any previous queens.
    """
    result = True
    for prev_row in range(row_q):
        #same column
        if board[prev_row][col_q] == 'Q':
            result = False
        
        # same diagonal
        delta = abs(row_q - prev_row)
        if (col_q - delta >= 0 and board[prev_row][col_q - delta] == 'Q') or \
            (col_q + delta < n and board[prev_row][col_q + delta] == 'Q'):
            result = False
    return result
        
```

```python
def print_board(board):
    for item in board:
        print(item)
    print(30 * '=')
```

```python
def solve_n_queen(board: list, n: int, row: int):
    if row == n:
        return board # solution
    for col in range(n):
        if is_safe(board, n, row, col):
            board[row][col] = 'Q'
            #print_board(board)
            result = solve_n_queen(board, n, row + 1)
            if result is not None:
                return result # solution
            board[row][col] = ' ' # backtrack
    return None
```

```python
n = 18
board = [[' ' for _ in range(n)] for __ in range(n)]

print_board(solve_n_queen(board, n, 0))
```

## Linear Search

```python
from typing import Any

def lineal_search(array: list, x: Any) -> int:
    """This function represents the behavior of a linear
    search making an iteration element by element in the
    provided list.
    
    Args:
        array(list): Set of items 
        x(Any): Element to be searched

    Returns:
        An integer with the index of the element. If element 
        is not found, -1 is returned as a logical mark.    
    """
    index = -1
    for i, elem in enumerate(array):
        if elem == x:
            index = i
            break
    return index

print(lineal_search([1,3,4,5,6,5,7,8,0], 0))
```

## Binary Search

```python
from typing import Any

def binary_search(a: list, x: Any, low: int, high: int) -> int:
    """This function represents the behavior of a simple
    binary search algorithm.
    
    Args:
        a(list): Sorted list of elements
        x(Any): element to be searched
        low(int): Lower index in the search space
        high(int): Higher index in the search space

    Returns:
        An integer with the index where the element is found.
        If not element, -1 is returned as a logical mark.
    """
    result = -1
    while low <= high:
        mid = (low + high) // 2
        if a[mid] == x:
            result = mid
            break
        elif a[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return result
```

```python
def catalog_search(catalog: list, isbn: int) -> int:
    """This function represents the behavior of a simple
    binary search algorithm to found books based on the ISBN.
    
    Args:
        catalog(list): Serted books catalog 
        isbn(int): book to be searched
    
    Returns:
        An integer with the index where the element is found.
        If not element, -1 is returned as a logical mark.
    """
    result = "Book not found."
    low = 0
    high = len(catalog) - 1

    while low <= high:
        mid = (low + high) // 2
        if catalog[mid]['isbn'] == isbn:
            result = catalog[mid]
            break
        elif catalog[mid]['isbn'] < isbn:
            low = mid + 1
        else:
            high = mid - 1
    return result
```

## Interpolation Search

```python
from typing import Any

def interpolation_search(a: list, x: Any, low: int, high: int) -> int:
    """This function represents the behavior of a simple
    interpolation search algorithm.
    
    Args:
        a(list): Sorted list of elements
        x(Any): element to be searched
        low(int): Lower index in the search space
        high(int): Higher index in the search space

    Returns:
        An integer with the index where the element is found.
        If not element, -1 is returned as a logical mark.
    """
    result = -1
    while low <= high and a[low] <= x <= a[high]:
        print(f'{low} --- {high}')
        pos = int(low + ( ((x - a[low]) / (a[high] - a[low])) * (high - low)))
        if a[pos] == x:
            result = pos
            break
        elif a[pos] < x:
            low = pos + 1
        else:
            high = pos - 1
    return result
```

```python
numbers = [i for i in range(5, 101, 5)]
print("Numbers:", numbers)
print( interpolation_search(numbers, 25, 0, len(numbers)-1) )
```

## Bubble Sort

```python
def bubble_sort(a: list):
    """This function represents the behavior of a basic
    bubble sort.
    
    Args:
        a(list): The array (all elements same type) to be sorted.

    Returns:
        The sorted array.
    """
    n = len(a)
    for i in range(n - 1):
        print(a)
        for j in range(n - i - 1):
            if a[j] > a[j+1]: # swap
                a[j], a[j+1] = a[j+1], a[j]
    return a
```
