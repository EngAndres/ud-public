# Euclidean GCD

This is an example of the __Euclidean Algorithm__ for this _course_.

```python
def euclidean_test(a: int, b: int) -> int:
    """
    This is a function to implement the classical euclidean algorithm
    to calculate the GCD.
    A should greater than B.

    Args:
        a(int): First number, should be greater than b
        b(int): Second number

    Returns:
        An integer with the GCD for the parameters provided.
    """
    if a <= b:
        a, b = b, a

    while b != 0:
        r = a % b
        a, b = b, r
    return a

result = euclidean_test(90, 60)
print(result)
```

```python
def euclidean_test_recursive(a: int, b: int) -> int:
    """
    This is a function to implement the classical euclidean algorithm
    to calculate the GCD.

    Args:
        a(int): First number, should be greater than b
        b(int): Second number

    Returns:
        An integer with the GCD for the parameters provided.
    """
    if b == 0:
        return a
    else:
        return euclidean_test_recursive(b, a % b)

print(euclidean_test_recursive(90, 60))
```

# Babylian Method for Square Root

```python
def square_root(s: int) -> float:
    """
    This method calculates a square root of an integer number based
    on the babylonean numeric method to approximate the root.

    Args:
        s(int): Integer as base to calculate the square root

    Returns:
        A float with the square root.
    """
    DIFF = 0.001
    x = s / 2

    while True:
        x_next = (1 / 2) * (x + (s / x))
        diff_x = abs(x - x_next)
        x = x_next
        if diff_x < DIFF:
            break
    return x

print(square_root(234123456789567678934567))
```

# FIND MAXIMUM IN AN ARRAY

```python
def find_maximum(a: list) -> int:
    """
    This method traverses an array and return the maximum item value.
    If the array is empty, and error message is returned.

    Args:
        a(list[int]): Array to be traversed

    Returns:
        An integer with the maximum value in the array
    """
    n = len(a)
    if n == 0:
        raise ValueError("Empty array is not valid.")
    
    max = a[0]
    for i in range(1, n):
        if a[i] > max:
            max = a[i]
    return max

print(max([11,13,4,13,5,6,6,13,4,2,8,4,3,9,13]))
```

## Find Minimum Coins for Change

```python
def get_coins_change(n: int, denominations: list) -> list:
    denominations = sorted(denominations, reverse=True)
    result = []
    for d in denominations:
        if d <= n:
            coins = n // d
            n = n % d
            result.append((d, coins))
    return result

denominations = [50, 1000, 500, 100, 200]
print(get_coins_change(29850, denominations))
```

```python
def knapsack(capacity, items):
    items = sorted(items, key=lambda item:item[1] / item[0], reverse=True)
    result = []
    for item in items:
        if item[0] <= capacity:
            result.append(item)
            capacity -= item[0]
    print(f'La maleta tiene {len(result)} con un valor de\
          { sum( [item[1] for item in result] ) }')
    return result

print(knapsack(150, items))
```

## Brutal Force - Cracking Password

```python
def get_possibilities(character_set: list, lenght: int, current: str = ""):
    if lenght == 1: # base case
        return [current + c for c in character_set]
    else:
        result = []
        for c in character_set:
            result.extend( get_possibilities(character_set, lenght - 1, current + c) )
        return result
```

```python
def get_password(character_set: list, password: str, max_length: int) -> str:
    result = None
    for length in range(1, int(max_length) + 1):
        possibilities = get_possibilities(character_set, length)
        for candidate in possibilities:
            if candidate == password:
                print("Found")
                result = candidate
    return result
            
character_set = list('abcdefghijklmnopqrstuvwxyz')
max_ = 6
password = 'zefhuj'
get_password(character_set, password, max_)
```

## N-Queens Problem

```python
def is_safe(board: list, n: int, row: int, col: int) -> bool:
    result = True
    for prev_row in range(row):
        # same column
        if board[prev_row][col] == 'Q':
            result = False
        # same diagonal
        delta = abs(row - prev_row)
        if (col - delta >= 0 and board[prev_row][col - delta] == 'Q') or \
            (col + delta < n and board[prev_row][col + delta] == 'Q'):
            result = False
    return result
```

```python
def solve_n_queens(board: list, row: int, n: int):
    if row == n:
        return board # :)
    if row >= len(board):
        board.append(['' for _ in range(n)]) 
    for col in range(n):
        if is_safe(board, n, row, col):
            board[row][col] = 'Q'
            result = solve_n_queens(board, row + 1, n)
            if result is not None:
                return result
            board[row][col] = '' # backtrack
    return None # no column is a solution
```

## Linear Search Basic

```python
from typing import Any

def linear_search(a: list, x: Any) -> int:
    """This function represents the behavior of a linear search
    making a iterator over a set of data.
    
    Args:
        a(list): Set of elemnts
        x(Any): Element to be searched

    Returns:
        An integer number with the index where x appears. If not, 
        -1 is returned as a logic mark.
    """
    index = -1
    for i, item in enumerate(a):
        if item == x:
            index = i
            break
    return index

print( linear_search([76,897,23,3,2,2,2,2,5.3], 5) )
```

## Binary Search

```python
from typing import Any

def binary_search(a: list, x: Any, low: int, high: int) -> int:
    """This function represents thew behavior of a typical
    binary search using a loop.

    Args:
        a(list): Set of elements
        x(Any): Element to be searched 
        low(int): Lower index in the search space
        high(int): Upper index in the search space
    
    Returns:
        An integer with the index with the element is found.
        If not found, -1 is returned as a logical mark.
    """
    result = -1
    while low <= high:
        print(f'{low} --- {high}')
        mid = (low + high) // 2
        if a[mid] == x:
            result = mid
            break
        elif a[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return result
```

## Interpolation Search

```python
from typing import Any

def interpolation_search(a: list, x: Any, low: int, high: int) -> int:
    """This function represents thew behavior of a typical
    interpolation search using a loop.

    Args:
        a(list): Set of elements
        x(Any): Element to be searched 
        low(int): Lower index in the search space
        high(int): Upper index in the search space
    
    Returns:
        An integer with the index with the element is found.
        If not found, -1 is returned as a logical mark.
    """
    result = -1
    while low <= high and x >= a[low] and x <= a[high]:
        print(f'{low} --- {high}')
        pos = int(low + ( ((x - a[low]) / (a[high] - a[low])) * (high - low) ))

        if a[pos] == x:
            result = pos
            break
        elif a[pos] < x:
            low = pos + 1
        else:
            high = pos - 1
    return result
```

## Bubblesort


```python
def bubble_sort(a: list) -> list:
    n = len(a)
    for i in range(n -1):
        for j in range(n - 1 - i):
            if a[j] > a[j + 1]:
                # swap
                a[j], a[j + 1] = a[j + 1], a[j]
    return a
```

```python
a = [4,6,7,3,5,9,8,4,3]
print( bubble_sort(a.copy()) )
```

## Selection Sort

```python
def selection_sort(a: list) -> list:
    n = len(a)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n): 
            if a[j] < a[min_index]:
                min_index =j
        a[i], a[min_index] = a[min_index],  a[i]
    return a
```

## Insertion Sort


```python
def insertion_sort(a: list) -> list:
    n = len(a)
    for i in range(1, n):
        value = a[i]
        j = i - 1
        while j >= 0 and a[j] > value:
            a[j + 1] = a[j]
            j -= 1
        a[j + 1] = value
    return a
```

```python
def partition(a: list) -> tuple:
    index = len(a) - 1
    pivot = a[index]
    left_a = []
    right_a = []
    for i in range(len(a)):
        if i != index:
            if a[i] < pivot:
                left_a.append(a[i])
            else:
                right_a.append(a[i])
    return pivot, left_a, right_a

def quick_sort(a: list) -> list:
    if len(a) == 1:
        return a
    else:
        pivot, left_a, right_a = partition(a)
        if len(left_a) > 0:
            left_a = quick_sort(left_a)
        if len(right_a) > 0:
            right_a = quick_sort(right_a)
        return left_a + [pivot] + right_a
```

# MergeSort

```python
def merge(left: list, right: list) -> list:
    result = []
    i_left = 0
    i_right = 0
    while i_left < len(left) and i_right < len(right):
        if left[i_left] < right[i_right]:
            result.append(left[i_left])
            i_left += 1
        else:
            result.append(right[i_right])
            i_right += 1
    result.extend(left[i_left:])
    result.extend(right[i_right:])
    return result
```

```python
def merge_sort(a: list) -> list:
    if len(a) == 1: 
        return a
    else:
        mid = len(a) // 2
        left = merge_sort(a[:mid])
        right = merge_sort(a[mid:])
        return merge(left, right)
```

```python
row_1 = list(map(int, input().split()))
row_2 = list(map(int, input().split()))
row_3 = list(map(int, input().split()))
square = [row_1, row_2, row_3]

def is_valid(square, row):
    valid = 1
    sum_row = sum(square[row])
    for i in range(row):
        sum_previous = sum(square[i])
        if  sum_row > sum_previous:
            valid = 0
            break
        elif sum_row != sum_previous:
            valid = -1
            break
    return valid

def is_valid_diagonal(square):
    sum_row = sum(square[0])
    sum_diagonal = sum([square[i][i] for i in range(len(square))])
    return sum_row == sum_diagonal 

def board(square, row):
    if row == len(square):
        return square if is_valid_diagonal(square) else None
    else:
        for i in range(1, 100001):
            square[row][row] = i
            valid = is_valid(square, row)
            if valid == 1:
                result = board(square, row + 1)
                if result is not None:
                    return result
            square[row][row] = 0
            if valid == 0:
                break
            
        return None       
             

square = board(square, 0)
for i in range(3):
    for j in range(3):
        print(square[i][j], end=" ")
    print()
```
